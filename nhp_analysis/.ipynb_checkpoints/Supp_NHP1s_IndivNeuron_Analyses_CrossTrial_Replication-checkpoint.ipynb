# Set up data parameters

```python
## All sessions (both subjects included)
sessions = ['100706','100730','100804','100820','100827','100913','100921','101024','101122','101128',
            '101207','101217','110110_01','110115_01','100724','100731','100817','100823','100828',
            '100915','101008','101027','101123','101202','101209','110106','110110_02','110120','100725',
            '100802','100818','100824','100907','100917','101009','101028','101124','101203','101210',
            '110107_01','110111_01','110121','100726','100803','100819','100826','100910','100920','101023',
            '101030','101127','101206','101216','110107_02','110111_02']

datadir = '/projects3/TaskFCMech/data/nhpData/'
tmin = -4000 # in ms
tmax = 4000 # in ms, this was my own doing

# Load in monkeyIDs
idfile = datadir + 'monkeyToSessionID.csv'
monkeyTable = pd.read_csv(idfile,delimiter=',')

replication_sessions = 0
monkeyID = {}
for i in range(1, len(sessions)+1):
    if i < 10:
        sess_str = 'session_ ' + str(i)
        name_str = 'name_ '  + str(i)
    else:
        sess_str = 'session_' + str(i)
        name_str = 'name_'  + str(i)
    
    session = str(monkeyTable[sess_str][0])
    name = monkeyTable[name_str][0]

    if name=='paula':
        replication_sessions += 1
        
    if session=='110111_02.mat': session = '110111_02'
        
    monkeyID[session] = name
    
exploratory_sessions = len(sessions) - replication_sessions
print 'Number of sessions with exploratory subject:', exploratory_sessions 
print 'Number of sessions with replication subject', replication_sessions
    
regions = ['PFC', 'FEF', 'LIP', 'MT', 'IT', 'V4']
```

# 1.0 Load all sessions for exploratory subject

#### Load other meta-data associated with neurons + task info

#### Compute downsampled firing rate using 50ms bins and 10ms shift

```python
analysis = reload(analysis)
ff_rest = []
ff_task = []
mu_rest = []
mu_task = []
var_rest = []
var_task = []

area_ind_all = []
nospikeneurons_allsessions = []

sessioncount = 1
for session in sessions:
    # Sort exploratory v replication sessions by subject ID
    if monkeyID[session]!='paula': continue
    print 'Loading data for session', session, '|', sessioncount, '/', replication_sessions 
    
    taskInfo = pd.read_csv(datadir + session + '_trialInfoAllTasks.csv')
    mocolTrials = np.where(taskInfo['task']=='mocol')[0]
    mocolInfo = pd.read_csv(datadir + session + '_trialInfoMOCOL.csv')
    
    unitInfo = pd.read_csv(datadir + session + '_unitInfo.csv')
    area_ind = np.asarray(unitInfo.area.values)

    print('\tLoading data...')
    h5f = h5py.File(datadir + session + '.h5', 'r')
    # Columns for h5f are reversed between matlab and python
    # Output is time x neuron x trials
    print('\tBinning spikes...')
    spikes_binned = analysis.spikeBinning(h5f['sta'][mocolTrials,:,:].T,binSize=50,shiftSize=10)
    h5f.close()
        
#    # Calculate mean-evoked response, and remove it from each trial
#    sta_removed = np.zeros(spikes_binned.shape)
#    tasksByTrial = mocolInfo['rule'].astype(str)
#    staByTask = {}
#    for task in np.unique(tasksByTrial):
#        task_ind = np.where(tasksByTrial==task)[0]
#        staByTask[task] = np.mean(spikes_binned[:,task_ind],axis=1)
#        for ind in task_ind:
#            sta_removed[:,ind] = spikes_binned[:,ind] - staByTask[task]
 
    print('\tComputing fano factor...')
    ff_rest_percond = []
    ff_task_percond = []
    mu_rest_percond = []
    mu_task_percond = []
    var_rest_percond = []
    var_task_percond = []
    tasksByTrial = mocolInfo['rule'].astype(str)
    for condition in ['color','motion']:
        ind = np.where(tasksByTrial==condition)[0]
        cond_mocolInfo = mocolInfo.iloc[ind]
        tmp_rest, tmp_task, tmp_murest, tmp_mutask, tmp_varrest, tmp_vartask = analysis.computeNeuronwiseFF(spikes_binned[:,:,ind],cond_mocolInfo,normalize=False)
        ff_rest_percond.append(tmp_rest)
        ff_task_percond.append(tmp_task)
        mu_rest_percond.append(tmp_murest)
        mu_task_percond.append(tmp_mutask)
        var_rest_percond.append(tmp_varrest)
        var_task_percond.append(tmp_vartask)
    ff_rest_percond = np.mean(np.asarray(ff_rest_percond),axis=0)
    ff_task_percond = np.mean(np.asarray(ff_task_percond),axis=0)
    mu_rest_percond = np.mean(np.asarray(mu_rest_percond),axis=0)
    mu_task_percond = np.mean(np.asarray(mu_task_percond),axis=0)
    var_rest_percond = np.mean(np.asarray(var_rest_percond),axis=0)
    var_task_percond = np.mean(np.asarray(var_task_percond),axis=0)
    
    # Remove any neurons with 0 spikes for either condition (since ff would be undefined (mean of 0 spikes))
    nospikeneurons_rest = np.where(np.isnan(ff_rest_percond))[0] 
    nospikeneurons_task = np.where(np.isnan(ff_task_percond))[0] 
    nospikeneurons_all = np.hstack((nospikeneurons_rest, nospikeneurons_task))

    tmp_ff_rest = np.delete(ff_rest_percond,nospikeneurons_all)
    tmp_ff_task = np.delete(ff_task_percond,nospikeneurons_all)
    tmp_mu_rest = np.delete(mu_rest_percond,nospikeneurons_all)
    tmp_mu_task = np.delete(mu_task_percond,nospikeneurons_all)
    tmp_var_rest = np.delete(var_rest_percond,nospikeneurons_all)
    tmp_var_task = np.delete(var_task_percond,nospikeneurons_all)
    
    area_ind = np.delete(area_ind,nospikeneurons_all)
    
    area_ind_all.extend(area_ind)
    
    ff_rest.extend(tmp_ff_rest)
    ff_task.extend(tmp_ff_task)
    mu_rest.extend(tmp_mu_rest)
    mu_task.extend(tmp_mu_task)
    var_rest.extend(tmp_var_rest)
    var_task.extend(tmp_var_task)
    print '\tnospikeneurons_all:', len(nospikeneurons_all)
    print '\tarea_ind (after deletion):', len(area_ind)
    nospikeneurons_allsessions.append(nospikeneurons_all)
    
    sessioncount += 1
    
```

# Plot change in firing rates for each neuron from rest to task 

```python
regions = ['PFC', 'FEF', 'LIP', 'MT', 'IT', 'V4']
mu_rest = np.asarray(mu_rest)
mu_task = np.asarray(mu_task)
ff_rest = np.asarray(ff_rest)
ff_task = np.asarray(ff_task)
area_ind_all = np.asarray(area_ind_all)
#
mu_rest_ordered = []
mu_task_ordered = []
ff_rest_ordered = []
ff_task_ordered = []

xtick_key = []
xtick_val = []

i = 0
for area in regions:
    xtick_key.append(i)
    xtick_val.append(area)
    ind = np.where(area_ind_all==area)[0]
    #
    mu_rest_ordered.extend(-1*np.sort(-mu_rest[ind]))
    mu_task_ordered.extend(-1*np.sort(-mu_task[ind]))
    #
    ff_rest_ordered.extend(-1*np.sort(-ff_rest[ind]))
    ff_task_ordered.extend(-1*np.sort(-ff_task[ind]))
    
    i += len(ind)
    
    
    
mu_rest_ordered = np.asarray(mu_rest_ordered)
mu_task_ordered = np.asarray(mu_task_ordered)
ff_rest_ordered = np.asarray(ff_rest_ordered)
ff_task_ordered = np.asarray(ff_task_ordered)
    
mu_diff_ordered = mu_task_ordered - mu_rest_ordered
ff_diff_ordered = ff_task_ordered - ff_rest_ordered
```

# Plot data

```python
i = 0
for area in regions:
    ind = np.where(area_ind_all==area)[0]
    #
    tmp_mu_diff = mu_task[ind] - mu_rest[ind] 
    mu_diff_ordered = -1*np.sort(-tmp_mu_diff) # Sort from positive->negative
    #
    tmp_ff_diff = ff_task[ind] - ff_rest[ind] 
    ff_diff_ordered = -1*np.sort(-tmp_ff_diff) # Sort from positive->negative
    
    plt.figure(figsize=(7,5))
    plt.title('Task vs. rest | ' + area,fontsize=22)
    sns.barplot(x=np.arange(len(mu_diff_ordered)), y=mu_diff_ordered)
    plt.axvline(x=np.min(np.where(mu_diff_ordered<0)[0]),color='grey',linestyle='--')
    plt.xticks(np.arange(0,len(ind)+1,np.floor(len(ind)/5)),np.arange(0,len(ind)+1,np.floor(len(ind)/5),dtype=int)+1,fontsize=16);
    plt.yticks(fontsize=16);
    plt.ylabel('$\Delta$ firing rate (Hz)',fontsize=20)
    plt.xlabel('Neuron',fontsize=20)
    plt.tight_layout()
    plt.savefig('Supp_NHP1s_IndivNeuron_Replication_taskVrestChange_Rate_' + area + '.png',dpi=300)


    plt.figure(figsize=(7,5))
    plt.title('Task vs. rest | ' + area,fontsize=22)
    sns.barplot(x=np.arange(len(ff_diff_ordered)), y=ff_diff_ordered)
    plt.axvline(x=np.min(np.where(ff_diff_ordered<0)[0]),color='grey',linestyle='--')
    plt.xticks(np.arange(0,len(ind)+1,np.floor(len(ind)/5)),np.arange(0,len(ind)+1,np.floor(len(ind)/5),dtype=int)+1,fontsize=16);
    plt.yticks(fontsize=16);
    plt.ylabel('$\Delta$ fano factor',fontsize=20);
    plt.xlabel('Neuron',fontsize=20);
    plt.tight_layout()
    plt.savefig('Supp_NHP1s_IndivNeuron_Replication_taskVrestChange_FanoFactor_' + area + '.png',dpi=300)
```

# Plot distribution [of neurons] of FR and FF for rest to task transition

```python
i = 0
for area in regions:
    ind = np.where(area_ind_all==area)[0]
    #
    tmp_mu_task = mu_task[ind] 
    tmp_mu_rest = mu_rest[ind] 
    #
    tmp_ff_task = ff_task[ind] 
    tmp_ff_rest = ff_rest[ind] 
    
    plt.figure(figsize=(7,5))
    plt.title('Firing rate distributions | ' + area,fontsize=22)
    sns.distplot(tmp_mu_task,bins=20,kde=False,label='Task',color='#f08080')
    sns.distplot(tmp_mu_rest,bins=20,kde=False,label='Rest',color='#add8e6')
    plt.xticks(fontsize=16);
    plt.yticks(fontsize=16);
    plt.xlabel('Firing rate (Hz)',fontsize=22);
    plt.ylabel('Frequency',fontsize=22);
    plt.legend(prop={'size': 16})
    plt.tight_layout()
    plt.savefig('Supp_NHP1s_IndivNeuron_Replication_taskVrestDistributions_Rates_' + area + '.png',dpi=300)


    plt.figure(figsize=(7,5))
    plt.title('Fano factor distributions | ' + area,fontsize=22)
    sns.distplot(tmp_ff_task,bins=20,kde=False,label='Task',color='#f08080')
    sns.distplot(tmp_ff_rest,bins=20,kde=False,label='Rest',color='#add8e6')
    plt.xticks(fontsize=16);
    plt.yticks(fontsize=16);
    plt.xlabel('Fano factor',fontsize=22);
    plt.ylabel('Frequency',fontsize=22);
    plt.legend(prop={'size': 16})
    plt.tight_layout()
    plt.savefig('Supp_NHP1s_IndivNeuron_Replication_taskVrestDistributions_FanoFactor_' + area + '.png',dpi=300)


```
