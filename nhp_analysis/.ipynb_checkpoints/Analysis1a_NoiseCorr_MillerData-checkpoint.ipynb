# 0.0 Define functions for loading data

```python
def loadRestData(subj,gsr=True):
    gsrstring = 'GSR' if gsr else 'noGSR'
    datafile = '/projects3/TaskFCMech/data/hcppreprocessedmsmall/' + subj + '_glmOutput_data.h5' 
    h5f = h5py.File(datafile,'r')
    data = []
    for run in range(1,5):
        dataid = subj + '_rest' + str(run) + '_residuals_' + gsrstring
        data.extend(h5f[dataid][:].T)
    data = np.asarray(data).T
    h5f.close()
    return data

def loadTaskActivity(subj,gsr=True):
    gsrstring = 'GSR' if gsr else 'noGSR'
    datafile = '/projects3/TaskFCMech/data/hcppreprocessedmsmall/' + subj + '_glmOutput_data.h5' 
    h5f = h5py.File(datafile,'r')
    dataid = subj + '_task_canonical_taskBetas_' + gsrstring
    betas = h5f[dataid][:]
    h5f.close()
    return betas

def loadTaskResidualsFIR(subj,gsr=True):
    gsrstring = 'GSR' if gsr else 'noGSR'
    datafile = '/projects3/TaskFCMech/data/hcppreprocessedmsmall/' + subj + '_glmOutput_data.h5'
    h5f = h5py.File(datafile,'r')
    dataid = subj + '_task_FIR_residuals_' + gsrstring
    betas = h5f[dataid][:]
    h5f.close()
    return betas

def loadTaskResidualsCanonical(subj,gsr=True):
    gsrstring = 'GSR' if gsr else 'noGSR'
    datafile = '/projects3/TaskFCMech/data/hcppreprocessedmsmall/' + subj + '_glmOutput_data.h5' 
    h5f = h5py.File(datafile,'r')
    dataid = subj + '_task_canonical_residuals_' + gsrstring
    betas = h5f[dataid][:]
    h5f.close()
    return betas

def loadRestDataFIR(subj,gsr=True):
    gsrstring = 'GSR' if gsr else 'noGSR'
    datafile = '/projects3/TaskFCMech/data/hcppreprocessedmsmall/' + subj + '_glmOutput_data.h5' 
    h5f = h5py.File(datafile,'r')
    dataid = subj + '_restAll_withTaskRegFIR_residuals_' + gsrstring
    data = h5f[dataid][:]
    h5f.close()
    return data

def loadTaskTiming(subj):
    X = ga.loadTaskTiming(subj)
    taskRegs = X['taskRegressors'][X['tMask'],:].copy()
    taskRegs = taskRegs > 0.25
    return taskRegs
```

## Compute task-general FC (across all task conditions versus rest FC)

```python
taskGeneralFC = np.zeros((nParcels,nParcels,len(subjNums)))
restGeneralFC = np.zeros((nParcels,nParcels,len(subjNums)))
for s in range(len(subjNums)):
    taskGeneralTiming = np.sum(task_timing[:,:,s],axis=1)>0
#     taskGeneralTiming = task_timing[:,22,s]>0
    
    taskGeneralFC[:,:,s] = np.corrcoef(taskTS_FIR[:,taskGeneralTiming,s])
    np.fill_diagonal(taskGeneralFC[:,:,s],0)
    
    restGeneralFC[:,:,s] = np.corrcoef(restTS_FIR[:,taskGeneralTiming,s])
    np.fill_diagonal(restGeneralFC[:,:,s],0)
    

    
# Fz transform
taskGeneralFC = np.arctanh(taskGeneralFC)
restGeneralFC = np.arctanh(restGeneralFC)

groupStatsFC = np.zeros((nParcels,nParcels,3))
groupStatsFC[:,:,0], groupStatsFC[:,:,1] = stats.ttest_rel(taskGeneralFC,restGeneralFC,axis=2)

# FDR-correction
triu_ind = np.triu_indices(nParcels,k=1)
groupStatsFC[:,:,2][triu_ind] = mc.fdrcorrection0(groupStatsFC[:,:,1][triu_ind])[0]
groupStatsFC[:,:,2] = groupStatsFC[:,:,2] + groupStatsFC[:,:,2].T
np.fill_diagonal(groupStatsFC[:,:,1],1) # diagonal p-value should be
```

```python
nproc = 10
taskRestFCDiff = np.zeros((nParcels,nTasks,3)) # t-stat, p-value, q-value
taskFC = np.zeros((nParcels,nParcels,nTasks,len(subjNums))) # From task time series using FIR regressed data
restFC = np.zeros((nParcels,nParcels,nTasks,len(subjNums))) # From rest time series, but using FIR regressed rest timings
FC_change_stats = np.zeros((nParcels,nParcels,nTasks,3)) # p and q values

taskSD = np.zeros((nParcels,nTasks,len(subjNums))) # From task time series using FIR regressed data
restSD = np.zeros((nParcels,nTasks,len(subjNums))) # From rest time series, but using FIR regressed rest timings

def computeFC(data):
    fc = np.corrcoef(data)
    np.fill_diagonal(fc,0)
    fc = np.arctanh(fc)
    return fc

for task in range(nTasks):
    sig_rois = np.where(taskActivity_stats[:,task,2] < 0.05)[0]
    nonsig_rois = np.where(taskActivity_stats[:,task,2] > 0.05)[0]
    
    inputsRest = []
    inputsTask = []
    scount = 0
    for subj in subjNums:
        tasktime = task_timing[:,task,scount] # isolate task
        
        tmptask = stats.zscore(taskTS_FIR[:,:,scount],axis=1)
        tmprest = stats.zscore(restTS_FIR[:,:,scount],axis=1)
        
        taskSD[:,task,scount] = np.std(tmptask[:,tasktime],axis=1)
        restSD[:,task,scount] = np.std(tmprest[:,tasktime],axis=1)
        
        inputsRest.append(restTS_FIR[:,tasktime,scount])
        inputsTask.append(taskTS_FIR[:,tasktime,scount])
        
        scount += 1
        
    pool = mp.Pool(processes=nproc)
    resultsRest = pool.map_async(computeFC,inputsRest).get()
    pool.close()
    pool.join()
    
    pool = mp.Pool(processes=nproc)
    resultsTask = pool.map_async(computeFC,inputsTask).get()
    pool.close()
    pool.join()
    
    for s in range(len(subjNums)):
        restFC[:,:,task,s] = resultsRest[s]
        taskFC[:,:,task,s] = resultsTask[s]
    
    t, p = stats.ttest_rel(taskFC[:,:,task,:], restFC[:,:,task,:],axis=2)
    FC_change_stats[:,:,task,0] = np.tanh(np.mean(taskFC[:,:,task,:] - restFC[:,:,task,:],axis=2))
    FC_change_stats[:,:,task,1] = p
    np.fill_diagonal(FC_change_stats[:,:,task,1],1.0) # diagonal should have p=1.0
    
    # Perform FDR-correction on changed connections
    triu_ind = np.triu_indices(nParcels,k=1)
    FC_change_stats[:,:,task,2][triu_ind] = mc.fdrcorrection0(FC_change_stats[:,:,task,1][triu_ind])[0]
    FC_change_stats[:,:,task,2] = FC_change_stats[:,:,task,2] + FC_change_stats[:,:,task,2].T
    np.fill_diagonal(FC_change_stats[:,:,task,2],0) # Diagonal should have q=1.0

    
        
        
        
        
```

# 1.0 *Hypothesis*: From rest to task states, there will be more quenching of correlated variability (and overall variability)

#### Group-averaged analysis (using average difference across tasks)

```python
# Perform a Group-level analysis; average across all tasks
taskfc_avg = np.mean(np.mean(np.mean(taskFC,axis=2),axis=1),axis=0) # Compute average correlation for each subject
restfc_avg = np.mean(np.mean(np.mean(restFC,axis=2),axis=1),axis=0)
fcstats = np.zeros((3,)) # avg r, t, p
fcstats[0] = np.mean(taskfc_avg-restfc_avg)
fcstats[1],fcstats[2] = stats.ttest_rel(taskfc_avg,restfc_avg)
print 'Average correlation difference between task versus rest:', fcstats[0]
print 'Task vs. Rest average FC difference: t =', fcstats[1], '| p =', fcstats[2]

taskSD_avg = np.mean(np.mean(taskSD,axis=1),axis=0)
restSD_avg = np.mean(np.mean(restSD,axis=1),axis=0)
sdstats = np.zeros((3,))
sdstats[0] = np.mean(taskfc_avg-restfc_avg)
sdstats[1],sdstats[2] = stats.ttest_rel(taskSD_avg,restSD_avg)
print 'Average SD difference between task versus rest:', sdstats[0]
print 'Task vs. Rest average SD difference: t =', sdstats[1], '| p =', sdstats[2]

plt.figure(figsize=(15,5))
plt.subplot(121)
sns.swarmplot(data=[taskfc_avg,restfc_avg])
plt.xticks(np.arange(2),['Average Task FC', 'Average Rest FC'],fontsize=18, fontname='FreeSans');
plt.ylabel('Average FC (whole-brain)',fontsize=20, fontname='FreeSans');
plt.title('Average FC from Rest to Task\nAcross 24 tasks', fontsize=24, fontname='FreeSans');
                   
                   
plt.subplot(122)
sns.swarmplot(data=[taskSD_avg,restSD_avg])
plt.xticks(np.arange(2),['Average Task SD', 'Average Rest SD'],fontsize=18, fontname='FreeSans');
plt.ylabel('Average SD (whole-brain)',fontsize=20, fontname='FreeSans');
plt.title('Average SD from Rest to Task\nAcross 24 tasks', fontsize=24, fontname='FreeSans');
plt.tight_layout()
```

#### Group-averaged analysis (using average difference across tasks)

```python
fcstats = np.zeros((24,4)) # r, t, p, q
sdstats = np.zeros((24,4)) # sd, t, p, q

for task in range(nTasks):
    taskfctmp = taskFC[:,:,task,:]
    restfctmp = restFC[:,:,task,:]
    
    tasksdtmp = taskSD[:,task,:]
    restsdtmp = restSD[:,task,:]
    
    taskfcavg = np.mean(np.mean(taskfctmp,axis=0),axis=0)
    restfcavg = np.mean(np.mean(restfctmp,axis=0),axis=0)
    
    tasksdavg = np.mean(tasksdtmp,axis=0)
    restsdavg = np.mean(restsdtmp,axis=0)

    fcstats[task,0] = np.mean(taskfcavg-restfcavg)
    sdstats[task,0] = np.mean(tasksdavg-restsdavg)
    
    fcstats[task,1], fcstats[task,2] = stats.ttest_rel(taskfcavg,restfcavg)
    sdstats[task,1], sdstats[task,2] = stats.ttest_rel(tasksdavg,restsdavg)

fcstats[:,3] = mc.fdrcorrection0(fcstats[:,2])[0]
sdstats[:,3] = mc.fdrcorrection0(sdstats[:,2])[0]

plt.figure(figsize=(15,5))
plt.subplot(121)
plt.title('T-values for FC changes for\ntask versus rest states',fontsize=24,fontname='FreeSans')
plt.ylabel('T-statistics',fontsize=20,fontname='FreeSans')
plt.xlabel('Task condition',fontsize=20,fontname='FreeSans')
tmp = np.multiply(fcstats[:,0],fcstats[:,3])
sns.barplot(np.arange(nTasks),tmp)


plt.subplot(122)
plt.title('T-values for SD changes for\ntask versus rest states',fontsize=24,fontname='FreeSans')
plt.ylabel('T-statistics',fontsize=20,fontname='FreeSans')
plt.xlabel('Task condition',fontsize=20,fontname='FreeSans')
tmp = np.multiply(sdstats[:,0],sdstats[:,3])
sns.barplot(np.arange(nTasks),tmp)
plt.tight_layout()
```

# 2.1 *Hypothesis*: Active regions will quench correlated variability within-network more than inactive regions
#### Run for each task separately

```python
activeROIsTaskFC = np.zeros((nTasks,len(subjNums)))
inactiveROIsTaskFC = np.zeros((nTasks,len(subjNums)))
statsActiveVsInactiveFC = np.zeros((nTasks,4)) # avg, t, p, q

activeROIsTaskSD = np.zeros((nTasks,len(subjNums)))
inactiveROIsTaskSD = np.zeros((nTasks,len(subjNums)))
statsActiveVsInactiveSD = np.zeros((nTasks,4)) # avg, t, p, q


for task in range(nTasks):
#     sig_rois = np.where(taskActivity_stats[:,task,2]==1.0)[0]
    t = taskActivity_stats[:,task,0]
    ind = sorted(range(len(t)), key=lambda k: -1*np.abs(t[k]))
    # Take top 50% of 'evoked' regions
    active_rois = ind[:(nParcels/2)]
    inactive_rois = ind[-(nParcels/2):]
    
    taskfc_task = taskFC[:,:,task,:]
    restfc_task = restFC[:,:,task,:]
    
    tasksd_task = taskSD[:,task,:]
    restsd_task = restSD[:,task,:]
    
    tmpfc = np.zeros((len(active_rois),len(subjNums)))
    tmpsd = np.zeros((len(active_rois),len(subjNums)))
    i = 0
    for roi in active_rois:
        net_ind = np.where(networkdef==networkdef[roi])[0]
        tmpfc[i,:] = np.mean(taskfc_task[roi,:,:],axis=0) - np.mean(restfc_task[roi,:,:],axis=0)
        tmpsd[i,:] = tasksd_task[roi,:] - restsd_task[roi,:]
        i += 1
    activeROIsTaskFC[task,:] = np.mean(tmpfc,axis=0)
    activeROIsTaskSD[task,:] = np.mean(tmpsd,axis=0)
    
    tmpfc = np.zeros((len(inactive_rois),len(subjNums)))
    tmpsd = np.zeros((len(inactive_rois),len(subjNums)))
    i = 0
    for roi in inactive_rois:
        net_ind = np.where(networkdef==networkdef[roi])[0]
        tmpfc[i,:] = np.mean(taskfc_task[roi,:,:],axis=0) - np.mean(restfc_task[roi,:,:],axis=0)
        tmpsd[i,:] = tasksd_task[roi,:] - restsd_task[roi,:]
        i += 1
    inactiveROIsTaskFC[task,:] = np.mean(tmpfc,axis=0)
    inactiveROIsTaskSD[task,:] = np.mean(tmpsd,axis=0)
        
    t, p = stats.ttest_rel(activeROIsTaskFC[task,:],inactiveROIsTaskFC[task,:])
    p = p/2.0 if t<0 else 1.0-p/2.0
    statsActiveVsInactiveFC[task,0] = np.mean(activeROIsTaskFC[task,:] - inactiveROIsTaskFC[task,:])
    statsActiveVsInactiveFC[task,1] = t
    statsActiveVsInactiveFC[task,2] = p
    
    t, p = stats.ttest_rel(activeROIsTaskSD[task,:],inactiveROIsTaskSD[task,:])
    p = p/2.0 if t<0 else 1.0-p/2.0
    statsActiveVsInactiveSD[task,0] = np.mean(activeROIsTaskSD[task,:] - inactiveROIsTaskSD[task,:])
    statsActiveVsInactiveSD[task,1] = t
    statsActiveVsInactiveSD[task,2] = p

statsActiveVsInactiveFC[:,3] = mc.fdrcorrection0(statsActiveVsInactiveFC[:,2])[0]
print 'FC changes: Number of tasks with significant decorrelations for active regions:', np.sum(statsActiveVsInactiveFC[:,3])
print 'SD changes: Number of tasks with significant variability quenching for active regions:', np.sum(statsActiveVsInactiveSD[:,3])
 
    
```
