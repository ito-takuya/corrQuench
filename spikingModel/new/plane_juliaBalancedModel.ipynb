# Set parameters

```julia
# Dynamic parameters
taue  = 15.0
taui = 10.0



# Construct network and synaptic connections
Ne = 4000
Ni = 1000
Nepop = 80 # cluster sizer (for E-E connections)
# Connectivity probabilities
pee = .2
pei = .5 
pie = .5
pii = .5
ree = 2.5 # Clustering coef
K = Ne*pee # Average number of E-E connections
# Synaptic efficacies
jeeout = 10.0 / (taue * sqrt(K))
jeein = 1.9 * 10.0 / (taue * sqrt(K))
jie = 4.0 / (sqrt(K) * taui)
jei = -16.0 * 1.2 / (taue * sqrt(K))
jii = -16.0 / (taui * sqrt(K))

@time W = constructMatrices(jeeout, jeein, jie, jei, jii, Ne=Ne, Ni=Ni, Nepop=Nepop, pee=pee, pei=pei, pie=pie, pii=pii, ree=ree);
```

# Simulate spontaneous dynamics

```julia
function downsample(data,dt=.1)
    # First downsample to 1ms
    xs, time = size(data)
    data2 = zeros(xs,Int(time*dt))
    ms = 0
    for i=1:time
        ms += dt
        ms2 = Int(ceil(ms))
        for x=1:xs
            data2[x,ms2] += data[x,i]
        end
    end

    return data2

end

function binSpikes(data,binsize=100)
    x, y = size(data)
    bins = Int(y/binsize)
    data2 = zeros(x,bins)
    bincount = 1
    i = 1
    while i<=y 
        for n=1:x
            data2[n,bincount] += sum(data[n,i:(i+99)])
        end
        bincount += 1
        i += 100
    end
    return data2
    
end
```

```julia
# include("sim_tito2.jl")

time = 2000 #ms 
dt = 0.1 #ms 
vre = 0.
threshe = 1 
threshi = 1 

tauerise = 1 
tauedecay = 3 #formerly 3
tauirise = 1 
tauidecay = 2 #formerly 2

muemin = 1.1 
muemax = 1.2 
muimin = 1.0 
muimax = 1.05

refrace = 5 
refraci = 5 

Nstime = 120
Nstimi = 0
# Nstimi = 140 
stimstart = 0 
stimend = 2000

ntrials = 10

stimrange = range(-0.6,0.5,step=0.05)
var_bins = zeros(Ncells,length(stimrange))
mean_bins = zeros(Ncells,length(stimrange))
i = 1
for stimamp in stimrange
    outputfilename = "/projects3/TaskFCMech/data/results/spikingModel/balanced_simoutput_spontaneous.h5"
    
    if stimamp<0
        stims = abs(stimamp)
        Nstime = 0
        Nstimi = 80
    elseif stimamp>=0
        stims = abs(stimamp)
        Nstime = 80
        Nstimi = 0
    end

    Ne = 4000
    tmp = zeros(Ncells,20,ntrials)
    for trial=1:ntrials
        print("Stim", stimamp, " trial: ", trial, "\n")
        @time times_spont, ns_spont, Ne, Ncells, T, spikes_spont, synInput_spont = sim(W, Ne=Ne, Ni=Ni, T=time, dt=dt, 
                                                                                 vre=vre, threshe=threshe, threshi=threshi, taue=taue, taui=taui,
                                                                                 tauerise=tauerise, tauedecay=tauedecay, tauirise=tauirise, tauidecay=tauidecay,
                                                                                 muemin=muemin, muemax=muemax, muimin=muimin, muimax=muimax,
                                                                                 Nstime=Nstime, Nstimi=Nstimi, stimamp=stims, stimstart=stimstart, stimend=stimend,
                                                                                 refrace=refrace, refraci=refraci, save=false, filename=outputfilename, overwrite=false);
        
        tmp[:,:,trial] = binSpikes(spikes_spont);
    
    end
    var_bins[:,i] = mean(var(tmp,dims=3),dims=2)
    mean_bins[:,i] = mean(mean(tmp,dims=3),dims=2)
    
    i += 1
    
    # Calculate 
end
# Print out mean firing rates
println("mean excitatory firing rate: ",mean(1000*ns_spont[1:Ne]/T)," Hz")
println("mean inhibitory firing rate: ",mean(1000*ns_spont[(Ne+1):Ncells]/T)," Hz")
```

```julia
# Create second plot
figure(figsize=(7,5))
title("Spike rate v. stimulation amplitude",fontsize=24)
# plot(mean(mean_bins,dims=1))
plot(stimrange,mean(mean_bins[1:Ne,:],dims=1)[:])
# plot(range(1,length(stimrange),step=1),mean(mean_bins,dims=1))
ylabel("Rate",fontsize=18)
xlabel("Stimulation amplitude",fontsize=18)
tight_layout()
# savefig("firingrate.png",dpi=150)

# Create second plot
figure(figsize=(7,5))
title("Spike variability v. stimulation amplitude",fontsize=24)
plot(stimrange,mean(var_bins[1:Ne,:],dims=1)[:])
ylabel("Variability",fontsize=18)
xlabel("Stimulation amplitude",fontsize=18)
tight_layout()
# savefig("synapticinput.png",dpi=150)
```
