```python
import numpy as np
from importlib import reload
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import convolve2d
%matplotlib inline
import numpy.matlib as matlib
from statsmodels.distributions.empirical_distribution import ECDF
import time as timepkg
import h5py
import scipy.stats as stats
import pandas as pd
```

```python
def bindata(trialdata,binSize,shiftSize):
    tLength = trialdata.shape[0]
    
    downSampledData = []
    i = 0 
    while i < (tLength-binSize):
        downSampledData.append(np.mean(trialdata[i:(i+binSize)],axis=0))
        i += shiftSize

    return np.asarray(downSampledData)
```

#### Load in data

```python
binsize=50
shiftsize=10
dt = .1
T = 2000
ntrials = 40

df = {}
df['Stimulation'] = []
df['Variability'] = []
df['Spike rate'] = []

data = np.zeros((ntrials,T))

fr = []
fr_mean = []
var_mean = []

stimamprange = np.arange(-.1,0.11,.01)

for stim in stimamprange:
    fr = []
    
    if np.round(stim,2)==0.0:
        stim = 0.0
    else:
        stim = np.round(stim,2)
    
    for i in np.arange(ntrials):

        outputfilename = "/projects3/TaskFCMech/data/results/spikingModel/balanced_simoutput_stim" + str(stim) + "_trial" + str(i+1) + ".h5"
        h5f = h5py.File(outputfilename,'r')
#         frE = h5f['frE'][:]
#         frI = h5f['frI'][:]
        Ne = h5f['Ne'][()] # Return scalar
#         Ncells = h5f['Ncells'][()] # Return scalar
#         times_spont = h5f['times'][:]
#         ns_spont = h5f['ns'][:]
#         T = h5f['T'][()] # Return scalar
        spikes = h5f['spikes'][:].T
        h5f.close()

        data[i,:] = np.mean(spikes[:Ne,:],axis=0)

        fr.append(bindata(data[i,:],binsize,shiftsize))

    fr = np.asarray(fr)
    
    fr_mean.append(np.mean(np.mean(fr,0)))
    var_mean.append(np.mean(np.var(fr,0)))
    
    df['Variability'].extend(np.var(fr,0))
    df['Stimulation'].extend(np.repeat(stim,len(np.var(fr,0))))
    df['Spike rate'].extend(np.mean(fr,0))



#     # # Task versus rest
#     t_fr, p_fr = stats.ttest_rel(np.mean(fr_stm,0),np.mean(fr_iti,0))
#     t_var, p_var = stats.ttest_rel(np.var(fr_stm,0),np.var(fr_iti,0))
#     print("Task versus rest, firing rate:", t_fr, p_fr)
#     print("Task versus rest, variability:", t_var, p_var)
```

```python
df['Spike rate'][:] = (df['Spike rate'][:] - np.min(df['Spike rate'][:]))/(np.max(df['Spike rate']) - np.min(df['Spike rate'][:]))
```

# Load in spontaneous data

```python
stimamparray = []
meanfr = []
meanvar = []
binsize=100
shiftsize=20



outputfilename = "/projects3/TaskFCMech/data/results/spikingModel/balanced_simoutput_spontaneous.h5"
h5f = h5py.File(outputfilename,'r')
frE = h5f['frE'][:]
frI = h5f['frI'][:]
Ne = h5f['Ne'][()] # Return scalar
Ncells = h5f['Ncells'][()] # Return scalar
times_spont = h5f['times'][:]
ns_spont = h5f['ns'][:]
T = h5f['T'][()] # Return scalar
synInputs = h5f['synInputsEI'][:] # Return scalar
h5f.close()


frE_spont = bindata(frE,binsize,shiftsize)
synE_spont = np.mean(synInputs[:,:Ne],1)
synE_spont = bindata(synE_spont,binsize,shiftsize)

outputfilename = "/projects3/TaskFCMech/data/results/spikingModel/balanced_simoutput_evoked.h5"
h5f = h5py.File(outputfilename,'r')
frE = h5f['frE'][:]
frI = h5f['frI'][:]
Ne = h5f['Ne'][()] # Return scalar
Ncells = h5f['Ncells'][()] # Return scalar
times_spont = h5f['times'][:]
ns_spont = h5f['ns'][:]
T = h5f['T'][()] # Return scalar
synInputs = h5f['synInputsEI'][:] # Return scalar
h5f.close()


frE_evoke = bindata(frE,binsize,shiftsize)
synE_evoke = np.mean(synInputs[:,:Ne],1)
synE_evoke = bindata(synE_evoke,binsize,shiftsize)

    
```

#### Plot excitatory mean firing rate vs. excitatory mean input
