```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import matplotlib as mpl
```

# 1.0 Define functions

Let $x_1$, $x_2$, $x_3$ be neural mass models. Define $w_ij$ to be the weight from neuron $x_i$ to $x_j$.

To identify the rate of change for the firing rates for these neuronal populations, we have

$$\tau_1 \frac{dx_1}{dt} = -x_1 + f\Big( w_{11} x_1 + w_{21} x_2 + w_{31} x_3 + b_1 + s_1 \Big) \qquad \qquad (1)$$

$$\tau_2 \frac{dx_2}{dt} = -x_2 + f\Big( w_{12} x_1 + w_{22} x_2 + w_{32} x_3 + b_2 + s_2 \Big) \qquad \qquad (2)$$

$$\tau_3 \frac{dx_3}{dt} = -x_3 + f\Big( w_{13} x_1 + w_{23} x_2 + w_{33} x_3 + b_3 + s_3 \Big) \qquad \qquad (3)$$

Where $f(x)$ refers to the nonlinear function defined above, $\tau_i$ refers to the time constant of neuron $i$, and $s_i$ is the gain on neuron $i$.


# Need to fix below equations

### 1.1 - Phase Portraits and Nullclines for $x_1$ and $x_2$ 

Mathematical formula for $x_1$ nullcline:
$$ x_2 = \frac{f^{-1} ( -x_1 ) - w_{11} x_1 - b_1 - s_1}{w_{21}} $$

Mathematical formula for $x_2$ nullcline:
$$ x_1 = \frac{f^{-1} ( -x_2 ) - w_{22} x_2 - b_2 - s_2}{-w_{12}} $$

```python
def sigmoid(x):
#     return np.tanh(x)
#     return 1/(1+np.exp(-.5*x))
    return 1/(1+np.exp(-x))

# def sigmoid(x):
#     zero_ind = np.where(x<=0)[0]
#     x[zero_ind] = 0
#     return x

def invsigmoid(x):
    return 2*np.log((x)/(1-(x)))

def runSim(W,D,B,S,t,xinit,tau,dt=.1):
    I = np.zeros((W.shape[0],t))
    # Set initial conditions
    I[:,0] = xinit
    
    for i in range(1,t):
        ik1 = -I[:,i-1] + sigmoid(np.dot(W,I[:,i-1]) + B + D[:,i-1] + S[:,i-1]) 
        ik1 = ik1/tau
        
        ia = I[:,i-1] + ik1*dt
                
        ik2 = -ia + sigmoid(np.dot(W,ia) + B + D[:,i] + S[:,i])
        ik2 = ik2/tau
        
        I[:,i] = I[:,i-1] + dt*((ik1+ik2)/2.0)
        
    return I[:,100:]

## Start editing from here

def findFP(w11,w12,w22,w21,b1,b2,s1,s2,t,xinit1,xinit2,tau1,tau2,dt=.1):
    """
    Finds the fixed point of the model for a particular parameter regime
    """
    I1 = np.zeros((t,))
    I2 = np.zeros((t,))
    # Set initial conditions
    I1[0] = xinit1
    I2[0] = xinit2
    
    for i in range(1,t):
        i1k1 = -I1[i-1] + sigmoid(w11*I1[i-1]+w21*I2[i-1]+b1+d1[i-1]+s1[i-1]) 
        i1k1 = i1k1/tau1
        i2k1 = -I2[i-1] + sigmoid(w22*I2[i-1]+w12*I1[i-1]+b2+d2[i-1]+s2[i-1]) 
        i2k1 = i2k1/tau2
        
        i1a = I1[i-1] + i1k1*dt
        i2a = I2[i-1] + i2k1*dt
        
        i1k2 = -i1a + sigmoid(w11*i1a+w21*i2a+b1+d1[i]+s1[i])
        i1k2 = i1k2/tau1
        i2k2 = -i2a + sigmoid(w22*i2a+w12*i1a+b2+d2[i]+s2[i]) 
        i2k2 = i2k2/tau2
        
        I1[i] = I1[i-1] + dt*((i1k1+i1k2)/2)
        I2[i] = I2[i-1] + dt*((i2k1+i2k2)/2)
        
        if I1[i] == I1[i-1] and I2[i] == I2[i-1]: 
            I1fp = I1[i] 
            I2fp = I2[i]
            break        
        
    return I1fp, I2fp




def nullsurface(W,D,I,B,S,ind=0):
    # 'ind' indicates which region's nullsurface to compute
    not_ind = np.where(np.arange(3)!=ind)[0]
    sorted_domain = np.arange(-1,1,.01)
    sorted_domain = np.arange(-0,1,.01)
    pos_ind = np.where(sorted_domain>0)[0]
    X, Y = np.meshgrid(sorted_domain[pos_ind], sorted_domain[pos_ind])
    
    invx = invsigmoid(X)
    nZ = (invx - W[ind,ind]*X - B[ind] - S[ind,0] - W[not_ind[0],ind]*Y)/W[not_ind[1],ind]
    nY = (invx - W[ind,ind]*X - B[ind] - S[ind,0] - W[not_ind[1],ind]*Y)/W[not_ind[0],ind]
    
    return X, nY, nZ
```

## 1.2 Construct connectivity matrices
#### A <-> B are strongly coupled
#### C is weakly coupled to both A and B

```python
nnodes = 3
W = np.ones((nnodes,nnodes))
akey = 0
bkey = 1
ckey = 2
# W[akey,bkey] = 5
# W[bkey,akey] = 5
# W[akey,ckey] = 2.5
# W[ckey,akey] = 2.5
# W[bkey,ckey] = 2.5
# W[ckey,bkey] = 2.5
W[akey,bkey] = 5
W[bkey,akey] = 5
W[akey,ckey] = 2.5
W[ckey,akey] = 2.5
W[bkey,ckey] = 2.5
W[ckey,bkey] = 2.5
np.fill_diagonal(W,1.0)
W = np.squeeze(W)

plt.figure()
ax = sns.heatmap(W, square=True,cmap='Reds',cbar=True)
ax.invert_yaxis()
plt.title('Connectivity matrix',fontsize=20,fontname='FreeSans')
plt.xlabel('Nodes',fontsize=18,fontname='FreeSans')
plt.xticks(range(3),['$x_1$','$x_2$','$x_3$'],fontsize=16,fontname='FreeSans')
plt.ylabel('Nodes',fontsize=18,fontname='FreeSans')
plt.yticks(range(3),['$x_1$','$x_2$','$x_3$'],fontsize=16,fontname='FreeSans')
# plt.colorbar()
plt.tight_layout()
# plt.savefig('Analysis4a_SynConnMat.jpg')
```

# 2.0 Run simulations

## 2.1 Default parameters (resting-state)

```python
# Set biases
B = np.ones((nnodes,))*(-5)

# Set time constants
tau = .1

# Set n timepoints
t = 5000


S = np.zeros((nnodes,t))
# S[:,:] = 0

# Set stimulation only to region 1
# s1[stimOnset:stimOffset] = 0

ntrials = 25
corrmat_rest = np.zeros((3,3,ntrials))
for i in range(ntrials):
    # Set initial conditions
    D = np.random.normal(0,1.0,(nnodes,t))
    xinit = np.random.normal(0,1.0,(nnodes,))
    I = runSim(W,D,B,S,t,xinit,tau)
    corrmat_rest[:,:,i] = np.corrcoef(I)
    np.fill_diagonal(corrmat_rest[:,:,i],0)

#####
# Generate example plot of last time series
plt.figure(figsize=(10,4))
# Plot
plt.plot(I.T)
# plt.plot(range(len(I2)),I2,label='Region2')
plt.legend()
plt.xlabel('Time',fontsize=20, fontname='Arial')
plt.ylabel('Activity',fontsize=20, fontname='Arial')
plt.title('Resting-state timeseries of three regions', fontsize=25,y=1.02, fontname='Arial')
# plt.savefig('Analysis4a_3DMod_RestingStateTS.jpg')


# xscalemin = -1
# xscalemax = 2
# yscalemin = -1
# yscalemax = 2
# zscalemin = -1
# zscalemax = 2
xscalemin = -.1
xscalemax = 1
yscalemin = -.1
yscalemax = 1
zscalemin = -.1
zscalemax = 1
vectorfield=False
fig = plt.figure(figsize=(10,7))
ax = Axes3D(fig)

ax.set_title('Phase-space: Spontaneous fluctuations', fontsize=25, y=1.02, fontname='FreeSans')
if vectorfield:
    Z, Y, X = np.mgrid[0:1:10j, 0:1:10j, 0:1:10j]
    X = X[::20]
    Y = Y[::20]
    Z = Z[::20]
#     Z, Y, X = np.mgrid[np.arange(yscalemin,yscalemax,.1), np.arange(xscalemin,xscalemax,.1), np.arange(zscalemin,zscalemax,.1)]
    V = (-X + sigmoid(W[0,0]*X+W[1,0]*Y+W[2,0]*Z+B[0]+S[0,0]))/tau
    U = (-Y + sigmoid(W[0,1]*X+W[1,1]*Y+W[2,1]*Z+B[1]+S[1,0]))/tau
    Ww = (-Z + sigmoid(W[0,2]*X+W[1,2]*Y+W[2,2]*Z+B[2]+S[2,0]))/tau
    ax.quiver(X, Y, Z, U, V, Ww, length=0.05)
    
# Plot trajectory
ax.plot3D(I[0,:], I[1,:], I[2,:])

X1, X2, X3 = nullsurface(W,D,I,B,S,ind=0)
ax.plot_surface(X1,X2,X3,alpha=.5,color='r',label='x1-nullsurface')

X2, X1, X3 = nullsurface(W,D,I,B,S,ind=1)
ax.plot_surface(X1,X2,X3, alpha=.5,color='b',label='x2-nullsurface')

X3, X1, X2 = nullsurface(W,D,I,B,S,ind=2)
ax.plot_surface(X1,X2,X3, alpha=.5,color='g',label='x3-nullsurface')

fake2Dline_x1 = mpl.lines.Line2D([0],[0], linestyle="none", c='r', marker = 'o')
fake2Dline_x2 = mpl.lines.Line2D([0],[0], linestyle="none", c='b', marker = 'o')
fake2Dline_x3 = mpl.lines.Line2D([0],[0], linestyle="none", c='g', marker = 'o')
leg = ax.legend([fake2Dline_x1, fake2Dline_x2, fake2Dline_x3],
          ['$x_1$-nullsurface', '$x_2$-nullsurface', '$x_3$-nullsurface'], 
          numpoints = 1)
bb = leg.get_bbox_to_anchor().inverse_transformed(ax.transAxes)
xOffset = 0.0
bb.x0 += xOffset
bb.x1 += xOffset
leg.set_bbox_to_anchor(bb, transform = ax.transAxes)

ax.set_xlim(xscalemin,xscalemax)
ax.set_ylim(yscalemin,yscalemax)
ax.set_zlim(zscalemin,zscalemax)
ax.set_xlabel('$X_1$',fontsize=16,fontname='FreeSans')
ax.set_ylabel('$X_2$',fontsize=16,fontname='FreeSans')
ax.set_zlabel('$X_3$',fontsize=16,fontname='FreeSans')
# fig.savefig('Analysis4a_3DMod_RestingStatePP.jpg')


plt.figure()
mat = np.mean(corrmat_rest,2)
ax = sns.heatmap(mat, square=True,cmap='Reds',cbar=True)
ax.invert_yaxis()
plt.title("Averaged rest correlation matrix",fontsize=20,fontname='FreeSans')
plt.xlabel('Nodes',fontsize=18,fontname='FreeSans')
plt.xticks(range(3),['$x_1$','$x_2$','$x_3$'],fontsize=16,fontname='FreeSans')
plt.ylabel('Nodes',fontsize=18,fontname='FreeSans')
plt.yticks(range(3),['$x_1$','$x_2$','$x_3$'],fontsize=16,fontname='FreeSans')
plt.tight_layout()
# plt.savefig('Analysis4a_3DMod_RestingStateFCMAT.jpg')


```

## 2.2 Stimulate both regions

```python
# Set biases
B = np.ones((nnodes,))*(-5)

# Set time constants
tau = .1

# Set n timepoints
t = 5000

S = np.zeros((nnodes,t))
# S[:,:] = 3
S[:,:] = 2.5



corrmat_task = np.zeros((3,3,ntrials))
for i in range(ntrials):
# Set initial conditions
    D = np.random.normal(0,1.0,(nnodes,t))
    xinit = np.random.normal(0,1.0,(nnodes,))

    #####
    ## Run simulations
    I = runSim(W,D,B,S,t,xinit,tau)
    # Calculate correlations
    corrmat_task[:,:,i] = np.corrcoef(I)
    np.fill_diagonal(corrmat_task[:,:,i],0)

#####
## Calculate nullclines
# I1_sort = np.arange(-5,100,.001)
# I2_sort = np.arange(-5,100,.001)

# d = 0
# nI1,xI1 = nullclineI1(w11,w21,d,I1_sort,b1,s1,b2)
# nI2,yI2 = nullclineI2(w22,w12,d,I2_sort,b2,s2,b1)

#####
# Generate plots
plt.figure(figsize=(10,4))
# Plot
plt.plot(I.T,)
# plt.plot(range(len(I2)),I2,label='Region2')
plt.legend()
plt.xlabel('Time',fontsize=20, fontname='Arial')
plt.ylabel('Activity',fontsize=20, fontname='Arial')
plt.title('Stimulus-evoked trial: timeseries of three regions', fontsize=25,y=1.02, fontname='Arial')
# plt.savefig('Analysis4a_3DMod_StimStateTS.jpg')


# xscalemin = -1
# xscalemax = 2
# yscalemin = -1
# yscalemax = 2
# zscalemin = -1
# zscalemax = 2
xscalemin = -.1
xscalemax = 1
yscalemin = -.1
yscalemax = 1
zscalemin = -.1
zscalemax = 1
vectorfield=False
fig = plt.figure(figsize=(10,7))
ax = Axes3D(fig)

ax.set_title('Phase-plane: Stimulus-evoked trial', fontsize=25, y=1.02, fontname='FreeSans')
if vectorfield:
    Z, Y, X = np.mgrid[0:1:10j, 0:1:10j, 0:1:10j]
    X = X[::20]
    Y = Y[::20]
    Z = Z[::20]
#     Z, Y, X = np.mgrid[np.arange(yscalemin,yscalemax,.1), np.arange(xscalemin,xscalemax,.1), np.arange(zscalemin,zscalemax,.1)]
    V = (-X + sigmoid(W[0,0]*X+W[1,0]*Y+W[2,0]*Z+B[0]+S[0,0]))/tau
    U = (-Y + sigmoid(W[0,1]*X+W[1,1]*Y+W[2,1]*Z+B[1]+S[1,0]))/tau
    Ww = (-Z + sigmoid(W[0,2]*X+W[1,2]*Y+W[2,2]*Z+B[2]+S[2,0]))/tau
    ax.quiver(X, Y, Z, U, V, Ww, length=0.05)
    
# Plot trajectory
ax.plot3D(I[0,:], I[1,:], I[2,:])

X1, X2, X3 = nullsurface(W,D,I,B,S,ind=0)
ax.plot_surface(X1,X2,X3,alpha=.5,color='r',label='x1-nullsurface')

X2, X1, X3 = nullsurface(W,D,I,B,S,ind=1)
ax.plot_surface(X1,X2,X3, alpha=.5,color='b',label='x2-nullsurface')

X3, X1, X2 = nullsurface(W,D,I,B,S,ind=2)
ax.plot_surface(X1,X2,X3, alpha=.5,color='g',label='x3-nullsurface')

fake2Dline_x1 = mpl.lines.Line2D([0],[0], linestyle="none", c='r', marker = 'o')
fake2Dline_x2 = mpl.lines.Line2D([0],[0], linestyle="none", c='b', marker = 'o')
fake2Dline_x3 = mpl.lines.Line2D([0],[0], linestyle="none", c='g', marker = 'o')
leg = ax.legend([fake2Dline_x1, fake2Dline_x2, fake2Dline_x3],
          ['$x_1$-nullsurface', '$x_2$-nullsurface', '$x_3$-nullsurface'], 
          numpoints = 1)
bb = leg.get_bbox_to_anchor().inverse_transformed(ax.transAxes)
xOffset = 0
bb.x0 += xOffset
bb.x1 += xOffset
leg.set_bbox_to_anchor(bb, transform = ax.transAxes)

ax.set_xlim(xscalemin,xscalemax)
ax.set_ylim(yscalemin,yscalemax)
ax.set_zlim(zscalemin,zscalemax)
ax.set_xlabel('$X_1$',fontsize=16,fontname='FreeSans')
ax.set_ylabel('$X_2$',fontsize=16,fontname='FreeSans')
ax.set_zlabel('$X_3$',fontsize=16,fontname='FreeSans')
# fig.savefig('Analysis4a_3DMod_StimStatePP.jpg')

plt.figure()
mat = np.mean(corrmat_task,2)
ax = sns.heatmap(mat, square=True,cmap='Reds',cbar=True)
ax.invert_yaxis()
plt.title("Averaged task correlation matrix",fontsize=20,fontname='FreeSans')
plt.xlabel('Nodes',fontsize=18,fontname='FreeSans')
plt.xticks(range(3),['$x_1$','$x_2$','$x_3$'],fontsize=16,fontname='FreeSans')
plt.ylabel('Nodes',fontsize=18,fontname='FreeSans')
plt.yticks(range(3),['$x_1$','$x_2$','$x_3$'],fontsize=16,fontname='FreeSans')
plt.tight_layout()
# plt.savefig('Analysis4a_3DMod_StimStateFCMAT.jpg')

```

## 2.3 Plot the difference between task and rest

```python
plt.figure(figsize=(14,7))
plt.subplot(121)
ax = sns.heatmap(np.mean(corrmat_rest,axis=2),square=True, cbar=True, cmap="Reds",cbar_kws={'fraction':.046})
ax.invert_yaxis()
plt.title('Intrinsic FC', fontsize=35,y=1.02, fontname='FreeSans')
plt.xlabel('Nodes',fontsize=35,fontname='FreeSans')
plt.xticks(np.arange(.5,3.5),['$X_1$','$X_2$','$X_3$'],fontsize=20,fontname='FreeSans')
plt.ylabel('Nodes',fontsize=35,fontname='FreeSans')
plt.yticks(np.arange(.5,3.5),['$X_1$','$X_2$','$X_3$'],fontsize=20,fontname='FreeSans')
plt.subplot(122)
ax = sns.heatmap(np.mean(corrmat_task,axis=2),square=True, cbar=True, cmap="Reds",cbar_kws={'fraction':.046})
ax.invert_yaxis()
plt.title('Stimulus-evoked FC', fontsize=35,y=1.02, fontname='FreeSans')
plt.xlabel('Nodes',fontsize=35,fontname='FreeSans')
plt.xticks(np.arange(.5,3.5),['$X_1$','$X_2$','$X_3$'],fontsize=20,fontname='FreeSans')
plt.ylabel('Nodes',fontsize=35,fontname='FreeSans')
plt.yticks(np.arange(.5,3.5),['$X_1$','$X_2$','$X_3$'],fontsize=20,fontname='FreeSans')
plt.tight_layout()
plt.savefig('Analysis4c_3DMode_RestAndTaskFC.pdf')

plt.figure(figsize=(7,7))
fcdiff, pvaldiff = stats.ttest_rel(corrmat_task,corrmat_rest,axis=2)
sigconn = pvaldiff*3.0 < 0.05
fcdiffsig = np.multiply(fcdiff,sigconn)
ax = sns.heatmap(fcdiffsig, square=True,cbar=True,center=0,cmap='coolwarm',cbar_kws={'fraction':.046})
ax.invert_yaxis()
plt.title('Stimulus-evoked vs\nintrinsic FC', fontsize=35,y=1.02, fontname='FreeSans')
plt.xlabel('Nodes',fontsize=35,fontname='FreeSans')
plt.xticks(np.arange(.5,3.5),['$X_1$','$X_2$','$X_3$'],fontsize=20,fontname='FreeSans')
plt.ylabel('Nodes',fontsize=35,fontname='FreeSans')
plt.yticks(np.arange(.5,3.5),['$X_1$','$X_2$','$X_3$'],fontsize=20,fontname='FreeSans')
plt.tight_layout()
plt.savefig('Analysis4c_3DMod_Task_vs_Rest_FCMAT.pdf')


```
