# Define threshold linear and SSN transfer functions

```python
def relu(x,bias):
    x = x + bias
    thresh = x > 0
    y = np.multiply(x,thresh)
    return y
    
def balancedSSN(x,bias):
    y = relu(x,-bias) - relu(x,bias)
    return y
    
```

## 2.4 Run statistical analysis of SD as a function of activity

```python
# Set connectivity weights
w11 = 1
# Set time constant
tau = .1
# Set n timepoints
t = 1000

stimAmps = [-3,0,3]
nSims = 10
sd_groupstats = np.zeros((len(stimAmps),nSims))
stimcount = 0
for s in stimAmps:
    for sim in range(nSims):
        d = np.random.normal(0,0.25,(t,))
        s1vec = np.ones((t,))*s
        
        ####
        # Calculate fixed point
        xinit1 = 0
        I1fp = findFP(w11,b,s,10000,xinit1,tau)    #####
        # Set initial conditions
        xinit = I1fp

        #####
        ## Run simulations
        I1 = runSim1D(w11,d,b,s1vec,t,xinit1,tau)
        # Calculate SD
        sd_groupstats[stimcount,sim] = np.std(I1)

    stimcount += 1

plt.figure()
sns.boxplot(data=sd_groupstats.T)
plt.title('SD-levels for de-activated,\nbaseline, and activated states',
          fontname='FreeSans',fontsize=18);
plt.xlabel('States',fontname='FreeSans',fontsize=16);
plt.ylabel('Standard deviation (SD)',fontname='FreeSans',fontsize=16);
plt.xticks([0,1,2],['De-activated', 'Baseline', 'Activated'], fontname='FreeSans',fontsize=14);
plt.tight_layout()
# plt.savefig('Analysis0d_StatisticalTestOfSDAcrossStates.pdf')

t, p = stats.ttest_rel(sd_groupstats[1,:],sd_groupstats[0,:])
print 'T-test, Baseline vs. De-activated state'
print '\tt =', t, '| p =', p

t, p = stats.ttest_rel(sd_groupstats[1,:],sd_groupstats[2,:])
print 'T-test, Baseline vs. Activated state'
print '\tt =', t, '| p =', p

```

# 3.0 Run linear stability analysis on fixed points across a range of stimulations

```python
import pandas as pd
def linearStabilityAnalysis1D(xstar,w,b,s,tau):
    """
    Calculates the characteristic time scale given a point xstar for our 1 dimensional system
    """
    secondderiv = (-1 + dxdtSigmoid(w*xstar+b+s))/float(tau)
    characteristic_timescale = 1/np.abs(secondderiv)
    return characteristic_timescale

def dxdtSigmoid(x):
    return np.exp(-x)/((np.exp(-x)+1)**2)
```

```python
# Set connectivity weights
w11 = 1
# Set time constant
tau = .1
# Set n timepoints
t = 1000

stimAmps = np.arange(-5,5.01,.05)
sd_values = []
timescale_values = []
tmp3 = []
for s in stimAmps:
    d = np.random.normal(0,0.25,(t,))
    s1vec = np.ones((t,))*s
    ## Calculate fixed point
    xinit1 = 0
    I1fp = findFP(w11,b,s,10000,xinit1,tau)
    tmp3.append(I1fp)
    # Set initial conditions
    xinit = I1fp
    ## Run simulations
    I1 = runSim1D(w11,d,b,s1vec,t,xinit1,tau)
    # Calculate SD
    sd_values.append(np.std(I1))

    # Calculate characteristic time point
    timescale_values.append(linearStabilityAnalysis1D(I1fp,w11,b,s,tau))
    

plt.figure(figsize=(6,5));
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
tmp2 = {}
tmp2['Stimulation amplitude'] = stimAmps
tmp2['SD of time series'] = sd_values
tmp2 = pd.DataFrame(tmp2)
sns.regplot('Stimulation amplitude','SD of time series',tmp2,color='k',fit_reg=False);
plt.xlabel('Stimulation amplitude',fontname='FreeSans',fontsize=20);
plt.ylabel('Average SD of timeseries',fontname='FreeSans',fontsize=20);
plt.title('Average SD of time series vs.\nStimulation amplitude', 
          y=1.02,fontsize=25,fontname='FreeSans');
plt.tight_layout();
plt.savefig('Analysis0d_SD_vs_StimAmp.pdf')


plt.figure(figsize=(6,5));
tmp2 = {}
tmp2['Stimulation amplitude'] = stimAmps
tmp2['Characteristic time scale'] = timescale_values
tmp2 = pd.DataFrame(tmp2)
sns.regplot('Stimulation amplitude','Characteristic time scale',tmp2,color='k',fit_reg=False)
plt.xlabel('Stimulation amplitude',fontname='FreeSans',fontsize=20);
plt.ylabel('Characteristic time scale',fontname='FreeSans',fontsize=20);
plt.title('Characteristic time scale vs.\nStimulation amplitude', 
          y=1.02,fontsize=25,fontname='FreeSans');
plt.tight_layout();
plt.savefig('Analysis0d_CharTimeScale_vs_StimAmp.pdf')

plt.figure(figsize=(6,5));
tmp2 = {}
tmp2['SD of time series'] = sd_values
tmp2['Characteristic time scale'] = timescale_values
tmp2 = pd.DataFrame(tmp2)
sns.regplot('SD of time series','Characteristic time scale',tmp2,color='k',fit_reg=True)
plt.xlabel('SD of time series',fontname='FreeSans',fontsize=20);
plt.ylabel('Characteristic time scale',fontname='FreeSans',fontsize=20);
tmp_spearmanr = round(stats.spearmanr(sd_values,timescale_values)[0],3)
plt.title('Characteristic time scale vs.\nSD of time series', 
          y=1.02,fontsize=25,fontname='FreeSans');
plt.annotate(xy=[0,.095], s='Rank correlation = ' + str(tmp_spearmanr),fontname='FreeSans',fontsize=14)
plt.tight_layout();
# plt.savefig('Analysis0d_CharTimeScale_vs_SD.pdf')


```
